```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD, Adam
import pandas as pd
```

```python
np.random.seed(1)
X = 2*np.random.random(size = (1000,2))-1
Y = X[:,0]*X[:,1] >= 0
```

```python
plt.plot(X[:,0], X[:,1], 'x')
plt.show()
```

```python
colors = ['red','blue']

plt.scatter(X[:,0], X[:,1], c=Y, cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```

```python
model = Sequential([
    tf.keras.layers.Flatten(input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='tanh'), #relu, sigmoid, 
    tf.keras.layers.Dense(2, activation='tanh'),
    tf.keras.layers.Dense(2, activation='linear') #linear
])
```

```python
X[0]
```

```python
predictions = model(X[:1]).numpy()
predictions
```

```python
tf.nn.softmax(predictions).numpy()
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
loss_fn(Y[:1], predictions).numpy()
```

```python
model.compile(optimizer=SGD(3e-2), #'adam'
              loss=loss_fn,
              metrics=['accuracy'])
```

```python
hist = model.fit(X, Y, epochs=500)
```

```python
plt.plot(hist.history['accuracy'])
plt.show()
```

```python
np.random.seed(5)
X_test = 2*np.random.random(size = (100,2))-1
Y_test = X_test[:,0]*X_test[:,1] >= 0
```

```python
model.evaluate(X_test,  Y_test, verbose=2)
```
